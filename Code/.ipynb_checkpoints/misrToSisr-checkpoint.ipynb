```python
import glob
from PIL import Image, ImageSequence

import torch
import numpy as np
import matplotlib.pyplot as plt
torch.cuda.is_available()
```

```python
LR_names = glob.glob('../../lightfield_mini/train/LF/*')
LR_tiffs = [Image.open(file) for file in LR_names]

LR = []
for j in range(len(LR_tiffs)):
    image = []
    for i, page in enumerate(ImageSequence.Iterator(LR_tiffs[j])):
        image.append(page)
    LR.append(image)
```

```python
HR_names = glob.glob('../../lightfield_mini/train/GT/*')
HR = [Image.open(file) for file in HR_names]
```

```python
for i in range(len(LR_tiffs)):
    plt.imshow(LR[i][0])
    LR[i][0].save("000"+ str(i)+ ".jpg")
```
